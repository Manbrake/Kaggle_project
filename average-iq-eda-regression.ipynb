# EDA

## The "Average IQ" variable

```python
# The distribution of the average IQ
sns.histplot(x=df['Average IQ'], kde=True)
sns.set_style("darkgrid")
```

```python
# We look at the top 10 countries with the highest average IQ from the data
top_10_IQ = df.sort_values(by='Average IQ', ascending=False).head(10)
fig = px.bar(top_10_IQ, x='Country', y='Average IQ', title="Top 10 countries with the highest average IQ")
fig.show()
```

```python
# We look at the top 10 countries with the lowest average IQ from the data
top_10_IQ_inv = df.sort_values(by='Average IQ').head(10)
fig = px.bar(top_10_IQ_inv, x='Country', y='Average IQ', title="Top 10 countries with the lowest average IQ")
fig.show()
```

## The "Participants" variable

```python
# The top ten countries by participants
top_10_par = df.sort_values(by='Participants', ascending=False).head(10)
fig = px.bar(top_10_par, x='Country', y='Participants', title="Top 10 countries with the highest participants", 
             color_discrete_sequence=['green'])
fig.show()
```

```python
# The top 10 countries with the least participants
top_10_par_inv = df.sort_values(by='Participants').head(10)
fig = px.bar(top_10_par_inv, x='Country', y='Participants', title="Top 10 countries with the least participants", 
             color_discrete_sequence=['green'])
fig.show()
```

There are huge differences in the number of participants per country. 

```python
sns.histplot(x=df['Participants'], kde=True)
sns.set_style("darkgrid")
```

One question that arises is this: 
Wouldn't the number of participants in a country correlate with the average IQ of that country?

```python
fig = px.scatter(
    df, x='Average IQ', y='Participants', opacity=0.65,
    trendline='ols', trendline_color_override='darkblue'
)
fig.update_layout(
    width=800,  
    height=600  
)
fig.show()
```

With an R² of 0.134, the participant variable explains 13% of the variation in mean IQ.  
New data would therefore be needed, notably concerning GDP, education, the country's political stability, the type of test given, etc., to be able to draw more solid conclusions from this data set.
